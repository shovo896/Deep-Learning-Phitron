```python
def initialize_parameters():
       parameters={}
       parameters['w1']=torch.ones((2,2))*0.1
       parameters['b1']=torch.zeros((2,1))*0.1

       parameters['w2']=torch.ones((2,1))*0.1
       parameters['b2']=torch.zeros((1,1))*0.1
       return parameters


```

better way 

```python
def initialize_parameters(layer_dims):
       torch.manual_seed(3)
       parameters={}
       L=len(layer_dims)
       for l in range(1,L):
              parameters['w'+str(l)]=torch.ones((layer_dims[l-1],layer_dims[l]))*0.1
              parameters['b'+str(l)]=torch.ones((layer_dims[l],1))*0.1
       return parameters
```

forward propagation

```python
def linear_forward(A_prev,w,b):
       z=torch.matmul(w.T,A_prev)+b
       return z
```

```python
initialize_parameters([2,2,1])
```

```python
def L_layer_forward(x,parameters):
       A=x
       L=len(parameters)//2

       for l in range(1,L+1):
              A_prev=A 
              wl=parameters['w'+str(l)]
              bl=parameters['b'+str(l)]
              A=linear_forward(A_prev,wl,bl)
              return A,A_prev
```

```python
def bce_loss(y_hat,y):
       loss=-(y*torch.log(y_hat)+(1-y)*torch.log(1-y_hat))
       return loss
```

```python
x=torch.tensor(2)
y=torch.tensor(0)
```

```python
w=torch.tensor(1)
b=torch.tensor(0)
```

```python
def bce(y_hat,y):
       y_hat=torch.clamp(y_hat,1e-7,1-1e-7)
       loss=-(y*torch.log(y_hat)+(1-y)*torch.log(1-y_hat))
       return loss
```

```python
z=w*x+b 
y_pred=torch.sigmoid(z)
loss=bce_loss(y_pred,y)
```

```python
loss
```

```python
#derivatives 
```

```python
loss_dy_pred=(y_pred-y)/(y_pred*(1-y_pred))
dy_pred_dz=y_pred*(1-y_pred)
dz_dw=x
dz_db=1
```

```python
dl_dw=loss_dy_pred*dy_pred_dz*dz_dw
dl_db=loss_dy_pred*dy_pred_dz*dz_db
```

```python
dl_dw
dl_db
```
