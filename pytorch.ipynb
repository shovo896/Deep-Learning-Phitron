weight init

```python
def initialize_parameters():
       parameters={}
       parameters['w1']=torch.ones((2,2))*0.1
       parameters['b1']=torch.zeros((2,1))*0.1

       parameters['w2']=torch.ones((2,1))*0.1
       parameters['b2']=torch.zeros((1,1))*0.1
       return parameters


```

```python
initialize_parameters()
```

better way 

```python
def initialize_parameters(layer_dims):
       torch.manual_seed(3)
       parameters={}
       L=len(layer_dims)
       for l in range(1,L):
              parameters['w'+str(l)]=torch.ones((layer_dims[l-1],layer_dims[l]))*0.1
              parameters['b'+str(l)]=torch.ones((layer_dims[l],1))*0.1
       return parameters
```
