```python
import pandas as pd
df = pd.read_csv('IMDB Dataset.csv',encoding='latin-1',engine='python',on_bad_lines='skip')
```

```python
df['review'] = df['review'].str.replace('<br />', ' ', regex=False)
df.head()
```

```python
df['review']
```

```python
import re 
def removal_html_tags(text):
    pattern = re.compile(r'<.*?>')
    return pattern.sub(r'', text)
```

```python
removal_html_tags(df.loc[0, 'review'])
```

```python
df['review']=df['review'].apply(removal_html_tags)
df['review']
```

```python
def remove_url(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)
```
