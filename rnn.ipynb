```python
import torch 
import torch.nn as nn
```

```python
class RNN(nn.Module):
       def __init__(self):
              #register kora 
              super(RNN,self).__init__()
              self.rnn = nn.RNN(input_size=5,hidden_size=3,num_layers=2,batch_first=True,batch_size=3)
              self.fc=nn.Linear(3, 1)
              self.sigmoid=nn.Sigmoid()
       def forward(self,x):
              out,_=self.rnn(x)
              out=out[:,-1,:]
              out=self.fc(out)
              out=self.sigmoid(out)
              return out




```
