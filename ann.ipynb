numpy array to tensors 

````````python
x_train_tensor = torch.from_numpy(x_train).to(torch.float64)
x_test_tensor = torch.from_numpy(x_test).to(torch.float64)
y_train_tensor = torch.from_numpy(y_train).to(torch.float64).view(-1, 1)
y_test_tensor = torch.from_numpy(y_test).to(torch.float64).view(-1, 1)
````````

dataset and dataloader

````````python
from torch.utils.data import Dataset,DataLoader 
````````

````````python
class CustomDataset(Dataset):
    def __init__(self,features, labels):
        self.features = features
        self.labels = labels
    def __len__(self):
        return len(self.features)
    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx]
````````

````````python
train_dataset=CustomDataset(x_train_tensor,y_train_tensor)
test_dataset=CustomDataset(x_test_tensor,y_train_tensor)
````````

````````python
x_train.shape[1]
````````

````````python
# loader bolte train loader ar hoilo test loader 
train_loader=DataLoader(train_dataset,batch_size=32,shuffle=True)
test_loader=DataLoader(test_dataset,batch_size=32,shuffle=True)
````````

````````python
import torch.nn as nn
````````

````````python
class MySimpleNN(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear = nn.Linear(num_features, 1)
        self.sigmoid = nn.Sigmoid()
        self.criterion = nn.BCELoss()

    def forward(self, features):
        features = features.to(self.linear.weight.dtype)
        return self.sigmoid(self.linear(features))

    def loss_function(self, y_pred, y):
        y = y.to(y_pred.dtype)
        return self.criterion(y_pred, y)
````````

````````python
class MySimpleNN(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear = nn.Linear(num_features, 1)
        self.sigmoid = nn.Sigmoid()
        self.criterion = nn.BCELoss()

    def forward(self, features):
        features = features.to(self.linear.weight.dtype)
        return self.sigmoid(self.linear(features))

    def loss_function(self, y_pred, y):
        y = y.to(y_pred.dtype)
        return self.criterion(y_pred, y)
````````

````````python
epochs=25
````````

````````python
loss_function=nn.BCELoss()
loss_function
````````

````````python
# create model
learning_rate = 0.01
model = MySimpleNN(x_train_tensor.shape[1]).double()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

for epoch in range(epochs):
    for batch_feature, batch_label in train_loader:
        y_pred = model(batch_feature)  # shape: [batch_size, 1]
        loss = model.loss_function(y_pred, batch_label.reshape(-1, 1))

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print(f"Epoch: {epoch + 1}, Loss: {loss.item():.6f}")
````````

````````python
model.eval()
accuracy_list=[]
with torch.no_grad():
    for batch_feature, batch_label in test_loader:
        y_pred = model(batch_feature)  # shape: [batch_size, 1]
        y_pred = (y_pred > 0.9).to(torch.float64)
        accuracy = (y_pred == batch_label).float().mean().item()
        accuracy_list.append(accuracy)

print(f'Average Accuracy: {sum(accuracy_list)/len(accuracy_list)}')
````````
