```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2, random_state=42)
```

numpy array to tensors 

```python
x_train_tensor = torch.from_numpy(x_train).to(torch.float64)
x_test_tensor = torch.from_numpy(x_test).to(torch.float64)
y_train_tensor = torch.from_numpy(y_train).to(torch.float64).view(-1, 1)
y_test_tensor = torch.from_numpy(y_test).to(torch.float64).view(-1, 1)
```

```python
import torch.nn as nn
```

```python
class MySimpleNN(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear = nn.Linear(num_features, 1)
        self.sigmoid = nn.Sigmoid()
        self.criterion = nn.BCELoss()

    def forward(self, features):
        features = features.to(self.linear.weight.dtype)
        return self.sigmoid(self.linear(features))

    def loss_function(self, y_pred, y):
        y = y.to(y_pred.dtype)
        return self.criterion(y_pred, y)
```

```python
class MySimpleNN(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear = nn.Linear(num_features, 1)
        self.sigmoid = nn.Sigmoid()
        self.criterion = nn.BCELoss()

    def forward(self, features):
        features = features.to(self.linear.weight.dtype)
        return self.sigmoid(self.linear(features))

    def loss_function(self, y_pred, y):
        y = y.to(y_pred.dtype)
        return self.criterion(y_pred, y)
```

```python
epochs=25
```

```python
loss_function=nn.BCELoss()
```

```python
# create model
learning_rate = 0.01
model = MySimpleNN(x_train_tensor.shape[1]).double()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

# define loop
for epoch in range(epochs):
    # forward pass
    y_pred = model(x_train_tensor)

    # loss calculate
    loss = model.loss_function(y_pred, y_train_tensor)

    # backward + update
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # print loss in each epoch
    print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```
