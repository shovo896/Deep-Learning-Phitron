build vocab 

```python
vocab={'<UNK>':0}
vocab
```

```python
for token in Counter(tokens).keys():
    print(token)
```

```python
for token in Counter(tokens).keys():
    if token not in vocab:
        vocab[token]=len(vocab)
```

extract sentences from data 

```python
input_sentences =document.split('\n')
input_sentences
```

```python
input_sentences[2]
```

```python
def text_to_indices(sentence, vocab):
    numerical_sentence=[]
    for token in word_tokenize(sentence):
        numerical_sentence.append(vocab.get(token, vocab['<UNK>']))
    return numerical_sentence
```

```python
input_numerical_sentence=[]
for sentence in input_sentences:
    input_numerical_sentence.append(text_to_indices(sentence.lower(), vocab))
```

```python
input_numerical_sentence
```

```python
training_sequence=[]

for sentence in input_numerical_sentence:
    for i in range(1, len(sentence)):
        training_sequence.append(sentence[:i+1])
```

```python
print(training_sequence)
```

```python
len_list=[]

for sequence in training_sequence:
   len_list.append(len(sequence))
```
