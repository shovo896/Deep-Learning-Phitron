```python
def text_to_indices(sentence, vocab):
    numerical_sentence=[]
    for token in word_tokenize(sentence):
        numerical_sentence.append(vocab.get(token, vocab['<UNK>']))
    return numerical_sentence
```

```python
input_numerical_sentence=[]
for sentence in input_sentences:
    input_numerical_sentence.append(text_to_indices(sentence.lower(), vocab))
```

```python
class CustomDataset(Dataset):
    def __init__(self, X, y):
        self.X=X
        self.y=y
    
    def __len__(self):
        return self.X.shape[0]
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

LSTM cell 

```python
class LSTMModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim):
        super(LSTMModel, self).__init__()
        self.embedding=nn.Embedding(vocab_size, embedding_dim)
        self.lstm=nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
        self.fc=nn.Linear(hidden_dim, vocab_size)
    
    def forward(self,x):
        x=self.embedding(x)
        output,(h_n,c_n)=self.lstm(x)
        output=self.fc(output[:,-1,:])
        return output
```

```python
model=LSTMModel(vocab_size=len(vocab), embedding_dim=50, hidden_dim=100)
model
```

```python
epochs=10
criterion=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model.parameters(), lr=0.001)
learn_rate=0.001
```

```python
for epoch in range(epochs):
    total_loss=0
    for batch_x ,batch_y in dataloader:
        batch_x, batch_y = batch_x.to(device), batch_y.to(device)
        optimizer.zero_grad()
        predictions = model(batch_x)
        loss = criterion(predictions, batch_y)
        loss.backward()
        optimizer.step()
        total_loss+=loss.item()
    print(f'Epoch {epoch+1}/{epochs}, Loss: {total_loss/len(dataloader)}')
```
